__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
#변수 초기화
붕어빵_판매_가능 = False

#붕어빵 재고와 주문 개수 비교하여 판매 가능 여부 확인
for i in 붕어빵_재고.keys():
    if 붕어빵_재고[i] < 붕어빵_주문_내역[i]:
        print(f"{i}붕어빵의 재고가 부족합니다.")
        붕어빵_판매_가능 = False
        break
    else:
        붕어빵_재고[i] -= 붕어빵_주문_내역[i]
        붕어빵_판매_가능 = True

#재고가 남아있으면 판매 메세지 출력
if 붕어빵_판매_가능:
    print(f"판매된 붕어빵 - 팥 : {붕어빵_주문_내역['팥']}개, 슈크림 : {붕어빵_주문_내역['슈크림']}개, 초코 : {붕어빵_주문_내역['초코']}개")
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
#변수 초기화
붕어빵_재고 = {"팥": 10, "슈크림": 8, "초코": 5}
주문_종료 = False

while(not 주문_종료):
    #변수 초기화
    붕어빵_판매_가능 = False

    #재고 출력
    print(f"현재 재고 - 팥 : {붕어빵_재고['팥']}개, 슈크림 : {붕어빵_재고['슈크림']}개, 초코 : {붕어빵_재고['초코']}개")
    print("----")

    #주문 입력
    붕어빵_주문 = input("주문할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 숫자만 입력하세요.(공백으로 구분. 예:1 0 2 )\n 주문을 종료할 경우 종료를 입력하세요. : ").split(" ")
    print("----")

    #주문, 종료 여부 확인
    if 붕어빵_주문[0] == "종료":
        주문_종료 = True
        print("주문이 종료되었습니다.")
        break
    else:
        #주문 입력값 저장
        붕어빵_주문_내역 = {"팥": int(붕어빵_주문[0]), "슈크림": int(붕어빵_주문[1]), "초코": int(붕어빵_주문[2])}

        #주문 내역 출력
        print(f"주문 내역 - 팥 : {붕어빵_주문_내역['팥']}개, 슈크림 : {붕어빵_주문_내역['슈크림']}개, 초코 : {붕어빵_주문_내역['초코']}개")
        print("----")

        #붕어빵 재고와 주문 개수 비교하여 판매 가능 여부 확인
        for i in 붕어빵_재고.keys():
            if 붕어빵_재고[i] < 붕어빵_주문_내역[i]:
                붕어빵_판매_가능 = False
                break
            else:
                붕어빵_판매_가능 = True

        #재고가 남아 있을 시 판매
        if 붕어빵_판매_가능:
            for i in 붕어빵_재고.keys():
                    붕어빵_재고[i] -= 붕어빵_주문_내역[i]
        else:
            print(f"{i}붕어빵의 재고가 부족합니다.")
            print("----")
            continue


        #재고가 남아있으면 판매 메세지 출력
        if 붕어빵_판매_가능:
            print(f"판매된 붕어빵 - 팥 : {붕어빵_주문_내역['팥']}개, 슈크림 : {붕어빵_주문_내역['슈크림']}개, 초코 : {붕어빵_주문_내역['초코']}개")
            print("----")
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
#변수 초기화
붕어빵_재고 = {"팥": 10, "슈크림": 8, "초코": 5}
붕어빵_관리 = {"팥": 0, "슈크림": 0, "초코": 0}
주문_종료 = False
관리자_비밀번호 = "1234"

#---주문을 종료할 때까지 동작---
while(not 주문_종료):

    #변수 초기화
    붕어빵_판매_가능 = False

    #재고 출력
    print(f"현재 재고 - 팥 : {붕어빵_재고['팥']}개, 슈크림 : {붕어빵_재고['슈크림']}개, 초코 : {붕어빵_재고['초코']}개")
    print("-" * 50)

    #---메인 입력창(주문, 종료, 관리자 로그인)---
    붕어빵_주문 = input("주문할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 숫자만 입력하세요.(공백으로 구분. 예:1 0 2 )\n주문을 종료할 경우 종료를 입력하세요.\n관리자 모드에 접근하려면 관리자를 입력하세요.\n입력 : ").split(" ")
    print("-" * 50)

    #---종료---
    if 붕어빵_주문[0] == "종료":
        주문_종료 = True
        print("주문이 종료되었습니다.")
        print("-" * 50)
        break

    #---관리자 로그인---
    elif 붕어빵_주문[0] == "관리자":

        #비밀번호 입력
        관리자_비밀번호_입력 = input("관리자 비밀번호를 입력하세요. : ")

        #비밀번호 일치
        if 관리자_비밀번호_입력 == 관리자_비밀번호:

            #---관리자 입력창(재고 추가, 종료)---
            붕어빵_관리 = input("추가할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 숫자만 입력하세요.(공백으로 구분. 예:1 0 2 )\n종료할 경우 종료를 입력하세요.\n입력 : ").split(" ")
            print("-" * 50)

            #관리자 모드 종료1
            if 붕어빵_관리[0] == "종료":
                print("관리자 모드를 종료하고 처음 화면으로 돌아갑니다.")
                print("-" * 50)
                continue

            #재고 추가
            else:
                붕어빵_관리_추가 = {"팥": int(붕어빵_관리[0]), "슈크림": int(붕어빵_관리[1]), "초코": int(붕어빵_관리[2])}
                for i in 붕어빵_재고.keys():
                    붕어빵_재고[i] += 붕어빵_관리_추가[i]
                print(f'팥 붕어빵을 {붕어빵_관리_추가["팥"]}개, 슈크림 붕어빵을 {붕어빵_관리_추가["슈크림"]}개, 초코 붕어빵을 {붕어빵_관리_추가["초코"]}개 추가했습니다.')
                print("-" * 50)

        #비밀번호 불일치
        else:
            print("비밀번호가 일치하지 않습니다. 처음 화면으로 돌아갑니다.")
            print("-" * 50)
            continue

    #---주문---
    else:

        #주문 입력값 저장
        붕어빵_주문_내역 = {"팥": int(붕어빵_주문[0]), "슈크림": int(붕어빵_주문[1]), "초코": int(붕어빵_주문[2])}

        #주문 내역 출력
        print(f'주문 내역 - 팥 : {붕어빵_주문_내역["팥"]}개, 슈크림 : {붕어빵_주문_내역["슈크림"]}개, 초코 : {붕어빵_주문_내역["초코"]}개')
        print("-" * 50)

        #붕어빵 재고와 주문 개수 비교하여 판매 가능 여부 확인
        for i in 붕어빵_재고.keys():
            if 붕어빵_재고[i] < 붕어빵_주문_내역[i]:
                붕어빵_판매_가능 = False
                break
            else:
                붕어빵_판매_가능 = True

        #재고가 남아 있을 시 판매
        if 붕어빵_판매_가능:
            for i in 붕어빵_재고.keys():
                    붕어빵_재고[i] -= 붕어빵_주문_내역[i]
            print(f'판매된 붕어빵 - 팥 : {붕어빵_주문_내역["팥"]}개, 슈크림 : {붕어빵_주문_내역["슈크림"]}개, 초코 : {붕어빵_주문_내역["초코"]}개')
            print("-" * 50)
        else:
            print(f"붕어빵의 재고가 부족합니다.")
            print("-" * 50)
            continue
```

    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
#변수 초기화
붕어빵 = {
    "재고": {
        "팥": 10, "슈크림": 8, "초코": 5
    },
    "가격": {
        "팥": 1000, "슈크림": 1200, "초코": 1500
    },
    "추가": {
        "팥": 0, "슈크림": 0, "초코": 0
    },
    "판매_개수": {
        "팥": 0, "슈크림": 0, "초코": 0
    }
}
붕어빵_매출 = 0
주문_종료 = False
관리자_비밀번호 = "1234"

#---주문을 종료할 때까지 동작---
while(not 주문_종료):

    #변수 초기화
    붕어빵_판매_가능 = False

    #재고 출력
    print(f'현재 재고 - 팥 : {붕어빵["재고"]["팥"]}개, 슈크림 : {붕어빵["재고"]["슈크림"]}개, 초코 : {붕어빵["재고"]["초코"]}개')
    print("-" * 50)

    #---메인 입력창(주문, 종료, 관리자 로그인)---
    메인_입력 = input("주문할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 숫자만 입력하세요.(공백으로 구분. 예:1 0 2 )\n"
                      "주문을 종료할 경우 종료를 입력하세요.\n"
                      "관리자 모드에 접근하려면 관리자를 입력하세요.\n"
                      "입력 : ").split(" ")
    print("-" * 50)

    #---종료---
    if 메인_입력[0] == "종료":

        #while문 False로
        주문_종료 = True

        print("주문이 종료되었습니다.")
        print("-" * 50)
        break

    #---관리자 로그인---
    elif 메인_입력[0] == "관리자":

        #비밀번호 입력
        관리자_비밀번호_입력 = input("관리자 비밀번호를 입력하세요. : ")

        #비밀번호 일치
        if 관리자_비밀번호_입력 == 관리자_비밀번호:

            #---관리자 입력창(재고 추가,매출 확인, 종료)---
            관리자_입력 = input("추가할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 숫자만 입력하세요.(공백으로 구분. 예:1 0 2 )\n"
                                "종료할 경우 종료를 입력하세요.\n"
                                "매출을 확인하려면 매출을 입력하세요.\n"
                                "입력 : ").split(" ")
            print("-" * 50)

            #관리자 모드 종료1
            if 관리자_입력[0] == "종료":
                print("관리자 모드를 종료하고 처음 화면으로 돌아갑니다.")
                print("-" * 50)
                continue

            #매출 확인
            elif 관리자_입력[0] == "매출":
                print("오늘 판매한 붕어빵의 갯수는\n"
                      f'{붕어빵["판매_개수"]["팥"]}개, 슈크림 : {붕어빵["판매_개수"]["슈크림"]}개, 초코 : {붕어빵["판매_개수"]["초코"]}개 입니다.\n'
                      f'오늘의 매출은 {붕어빵_매출}원입니다.')
                print("-" * 50)
                continue

            #재고 추가
            else:
                붕어빵["추가"] = {"팥": int(관리자_입력[0]), "슈크림": int(관리자_입력[1]), "초코": int(관리자_입력[2])}
                for i in 붕어빵["재고"].keys():
                    붕어빵["재고"][i] += 붕어빵["추가"][i]
                print(f'팥 붕어빵을 {붕어빵["추가"]["팥"]}개, 슈크림 붕어빵을 {붕어빵["추가"]["슈크림"]}개, 초코 붕어빵을 {붕어빵["추가"]["초코"]}개 추가했습니다.')
                print("-" * 50)

        #비밀번호 불일치
        else:
            print("비밀번호가 일치하지 않습니다. 처음 화면으로 돌아갑니다.")
            print("-" * 50)
            continue

    #---주문---
    else:

        #주문 입력값 저장
        붕어빵["주문"] = {"팥": int(메인_입력[0]), "슈크림": int(메인_입력[1]), "초코": int(메인_입력[2])}

        #주문 내역 출력
        print(f'주문 내역 - 팥 : {붕어빵["주문"]["팥"]}개, 슈크림 : {붕어빵["주문"]["슈크림"]}개, 초코 : {붕어빵["주문"]["초코"]}개')
        print("-" * 50)

        #붕어빵 재고와 주문 개수 비교하여 판매 가능 여부 확인
        for i in 붕어빵["재고"].keys():
            if 붕어빵["재고"][i] < 붕어빵["주문"][i]:
                붕어빵_판매_가능 = False
                break
            else:
                붕어빵_판매_가능 = True

        #재고가 남아 있을 시 판매
        if 붕어빵_판매_가능:
            for i in 붕어빵["재고"].keys():
                    붕어빵["재고"][i] -= 붕어빵["주문"][i]
                    #판매 개수에 추가
                    붕어빵["판매_개수"][i] += 붕어빵["주문"][i]
                    #매출 계산
                    붕어빵_매출 += 붕어빵["가격"][i] * 붕어빵["판매_개수"][i]

            print(f'판매된 붕어빵 - 팥 : {붕어빵["주문"]["팥"]}개, 슈크림 : {붕어빵["주문"]["슈크림"]}개, 초코 : {붕어빵["주문"]["초코"]}개')
            print("-" * 50)

        #재고가 부족할 시 메인 화면으로 이동
        else:
            print(f"붕어빵의 재고가 부족합니다.")
            print("-" * 50)
            continue
print(붕어빵)
```
