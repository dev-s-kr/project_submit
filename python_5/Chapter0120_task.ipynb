```python
class 붕어빵판매점:

    def __init__(self):
        self.붕어빵 = {
            "재고": {
                "팥": 10, "슈크림": 8, "초코": 5
            },
            "가격": {
                "팥": 1000, "슈크림": 1200, "초코": 1500
            },
            "추가": {
                "팥": 0, "슈크림": 0, "초코": 0
            },
            "판매_개수": {
                "팥": 0, "슈크림": 0, "초코": 0
            }
        }
        self.붕어빵_매출 = 0
        self.주문_종료 = False
        self.붕어빵_판매_가능_여부 = False
        self.관리자_비밀번호 = "1234"

    def 메인_화면(self):
        while not self.주문_종료:
            self.재고_확인()
            메인_입력 = input(
            "주문할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 숫자만 입력하세요.(공백으로 구분. 예:1 0 2 )\n"
            "주문을 종료할 경우 종료를 입력하세요.\n"
            "관리자 모드에 접근하려면 관리자를 입력하세요.\n"
            "입력 : ").split(" ")
            print("-" * 50)
            try:
                if not (all(item.isdigit() for item in 메인_입력) or len(메인_입력) == 3 or 메인_입력[0] in ["종료", "관리자"]):
                    raise Exception("올바른 입력값을 입력해주세요.")
                self.주문_처리(메인_입력)
            except Exception as e:
                print("에러 발생 : ", e)
                print("-" * 50)

    def 재고_확인(self):
        print(f'현재 재고 - 팥 : {self.붕어빵["재고"]["팥"]}개, 슈크림 : {self.붕어빵["재고"]["슈크림"]}개, 초코 : {self.붕어빵["재고"]["초코"]}개')
        print("-" * 50)

    def 주문_처리(self, 메인_입력):
        if 메인_입력[0] == "종료":
            self.메인_입력_종료()
        elif 메인_입력[0] == "관리자":
            self.메인_입력_관리자(self.관리자_비밀번호)
        else:
            self.메인_입력_붕어빵(메인_입력)

    def 메인_입력_종료(self):
        self.주문_종료 = True
        print("주문이 종료되었습니다.")
        print("-" * 50)

    def 메인_입력_붕어빵(self, 메인_입력):
        self.붕어빵["주문"] = {"팥": int(메인_입력[0]), "슈크림": int(메인_입력[1]), "초코": int(메인_입력[2])}
        print(f'주문 내역 - 팥 : {self.붕어빵["주문"]["팥"]}개, 슈크림 : {self.붕어빵["주문"]["슈크림"]}개, 초코 : {self.붕어빵["주문"]["초코"]}개')
        print("-" * 50)
        self.주문_개수_재고_비교()

    def 주문_개수_재고_비교(self):
        for i in self.붕어빵["재고"].keys():
            if self.붕어빵["재고"][i] < self.붕어빵["주문"][i]:
                붕어빵_판매_가능_여부 = False
                break
            else:
                붕어빵_판매_가능_여부 = True
        if 붕어빵_판매_가능_여부 == True:
            self.붕어빵_재고_충분()
        else:
            self.붕어빵_재고_부족()
    def 붕어빵_재고_충분(self):
        for i in self.붕어빵["재고"].keys():
            self.붕어빵["재고"][i] -= self.붕어빵["주문"][i]
            self.붕어빵["판매_개수"][i] += self.붕어빵["주문"][i]
            self.붕어빵_매출 += self.붕어빵["가격"][i] * self.붕어빵["판매_개수"][i]

    def 붕어빵_재고_부족(self):
        print(f"붕어빵의 재고가 부족합니다.")
        print("-" * 50)

def main():
    붕어빵_판매_시스템 = 붕어빵판매점()
    while(not 붕어빵_판매_시스템.주문_종료):
        붕어빵_판매_시스템.메인_화면()

if __name__ == "__main__":
    main()메인_화면
```

__[과제 4: 관리자 모드 메서드 admin_mode]__

목표: admin_mode 메서드를 구현하여 관리자 모드에서 재고를 추가할 수 있도록 합니다.

설명: bread_type과 additional_stock을 입력받아 재고를 추가합니다. 잘못된 입력 값이 들어온 경우 적절한 메시지를 출력하세요.

```python
class 붕어빵판매점:

    def __init__(self):
        self.붕어빵 = {
            "재고": {
                "팥": 10, "슈크림": 8, "초코": 5
            },
            "가격": {
                "팥": 1000, "슈크림": 1200, "초코": 1500
            },
            "추가": {
                "팥": 0, "슈크림": 0, "초코": 0
            },
            "판매_개수": {
                "팥": 0, "슈크림": 0, "초코": 0
            }
        }
        self.붕어빵_매출 = 0
        self.주문_종료 = False
        self.관리자_입력_종료 = False
        self.붕어빵_판매_가능_여부 = False
        self.관리자_비밀번호 = "1234"

    def 메인_화면(self):
        while not self.주문_종료:
            self.재고_확인()
            메인_입력 = input(
                "주문할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 「숫자」만 입력하세요.(「공백으로 구분.」 예:1 0 2 )\n"
                "주문을 종료할 경우 「종료」를 입력하세요.\n"
                "관리자 모드에 접근하려면 「관리자」를 입력하세요.\n"
                "입력 : ").split(" ")
            print("-" * 50)
            try:
                if not (all(item.isdigit() for item in 메인_입력) or len(메인_입력) == 3 or 메인_입력[0] in ["종료", "관리자"]):
                    raise Exception("올바른 입력값을 입력해주세요.")
                self.주문_처리(메인_입력)
            except Exception as e:
                print("에러 발생 : ", e)
                print("-" * 50)

    def 재고_확인(self):
        print(f'현재 재고 - 팥 : {self.붕어빵["재고"]["팥"]}개, 슈크림 : {self.붕어빵["재고"]["슈크림"]}개, 초코 : {self.붕어빵["재고"]["초코"]}개')
        print("-" * 50)

    def 주문_처리(self, 메인_입력):
        if 메인_입력[0] == "종료":
            self.메인_입력_종료()
        elif 메인_입력[0] == "관리자":
            self.메인_입력_관리자()
        else:
            self.메인_입력_붕어빵(메인_입력)

    def 메인_입력_종료(self):
        self.주문_종료 = True
        print("주문이 종료되었습니다.")
        print("-" * 50)

    def 메인_입력_관리자(self):
        관리자_비밀번호_입력 = input("관리자 비밀번호를 입력하세요. : ")
        print("-" * 50)
        self.관리자_비밀번호_확인(관리자_비밀번호_입력)

    def 관리자_비밀번호_확인(self, 관리자_비밀번호_입력):
        if 관리자_비밀번호_입력 == self.관리자_비밀번호:
            self.관리자_입력_종료 = False
            self.관리자_화면()
        else:
            self.관리자_비밀번호_불일치()

    def 관리자_화면(self):
        while not self.관리자_입력_종료:
            self.재고_확인()
            관리자_입력 = input(
                "[ --- 관리자 모드 --- ]\n"
                "추가할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 「숫자만」 입력하세요.(「공백으로 구분.」 예:1 0 2 )\n"
                "종료할 경우 「종료」를 입력하세요.\n"
                "매출을 확인하려면 「매출」을 입력하세요.\n"
                "입력 : ").split(" ")
            print("-" * 50)
            try:
                if not (all(item.isdigit() for item in 관리자_입력) or len(관리자_입력) == 3 or 관리자_입력[0] in ["종료", "매출"]):
                    raise Exception("올바른 입력값을 입력해주세요.")
                self.관리자_입력_처리(관리자_입력)
            except Exception as e:
                print("에러 발생 : ", e)
                print("-" * 50)

    def 관리자_입력_처리(self, 관리자_입력):
        if 관리자_입력[0] == "종료":
            self.관리자_입력_종료 = True
            print("주문 화면으로 돌아갑니다.")
            print("-" * 50)
        else:
            self.관리자_입력_재고_추가(관리자_입력)


    def 관리자_입력_재고_추가(self, 관리자_입력):
        self.붕어빵["추가"] = {"팥": int(관리자_입력[0]), "슈크림": int(관리자_입력[1]), "초코": int(관리자_입력[2])}
        for i in self.붕어빵["재고"].keys():
            self.붕어빵["재고"][i] += self.붕어빵["추가"][i]
        print(f'팥 붕어빵을 {self.붕어빵["추가"]["팥"]}개, 슈크림 붕어빵을 {self.붕어빵["추가"]["슈크림"]}개, 초코 붕어빵을 {self.붕어빵["추가"]["초코"]}개 추가했습니다.')
        print("-" * 50)

    def 관리자_비밀번호_불일치(self):
        print("비밀번호가 일치하지 않습니다. 처음 화면으로 돌아갑니다.")
        print("-" * 50)

    def 메인_입력_붕어빵(self, 메인_입력):
        self.붕어빵["주문"] = {"팥": int(메인_입력[0]), "슈크림": int(메인_입력[1]), "초코": int(메인_입력[2])}
        print(f'주문 내역 - 팥 : {self.붕어빵["주문"]["팥"]}개, 슈크림 : {self.붕어빵["주문"]["슈크림"]}개, 초코 : {self.붕어빵["주문"]["초코"]}개')
        print("-" * 50)
        self.주문_개수_재고_비교()

    def 주문_개수_재고_비교(self):
        for i in self.붕어빵["재고"].keys():
            if self.붕어빵["재고"][i] < self.붕어빵["주문"][i]:
                붕어빵_판매_가능_여부 = False
                break
            else:
                붕어빵_판매_가능_여부 = True
        if 붕어빵_판매_가능_여부 == True:
            self.붕어빵_재고_충분()
        else:
            self.붕어빵_재고_부족()

    def 붕어빵_재고_충분(self):
        for i in self.붕어빵["재고"].keys():
            self.붕어빵["재고"][i] -= self.붕어빵["주문"][i]
            self.붕어빵["판매_개수"][i] += self.붕어빵["주문"][i]
            self.붕어빵_매출 += self.붕어빵["가격"][i] * self.붕어빵["판매_개수"][i]
        print(f"주문이 성공적으로 완료됐습니다.")
        print("-" * 50)

    def 붕어빵_재고_부족(self):
        print(f"붕어빵의 재고가 부족합니다.")
        print("-" * 50)

def main():
    붕어빵_판매_시스템 = 붕어빵판매점()
    붕어빵_판매_시스템.메인_화면()

if __name__ == "__main__":
    main()
```

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다.

```python
class 붕어빵판매점:

    def __init__(self):
        self.붕어빵 = {
            "재고": {
                "팥": 10, "슈크림": 8, "초코": 5
            },
            "가격": {
                "팥": 1000, "슈크림": 1200, "초코": 1500
            },
            "추가": {
                "팥": 0, "슈크림": 0, "초코": 0
            },
            "판매_개수": {
                "팥": 0, "슈크림": 0, "초코": 0
            }
        }
        self.붕어빵_매출 = 0
        self.주문_종료 = False
        self.관리자_입력_종료 = False
        self.붕어빵_판매_가능_여부 = False
        self.관리자_비밀번호 = "1234"

    def 메인_화면(self):
        while not self.주문_종료:
            self.재고_확인()
            메인_입력 = input(
                "주문할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 「숫자」만 입력하세요.(「공백으로 구분.」 예:1 0 2 )\n"
                "주문을 종료할 경우 「종료」를 입력하세요.\n"
                "관리자 모드에 접근하려면 「관리자」를 입력하세요.\n"
                "입력 : ").split(" ")
            print("-" * 50)
            try:
                if not (all(item.isdigit() for item in 메인_입력) or len(메인_입력) == 3 or 메인_입력[0] in ["종료", "관리자"]):
                    raise Exception("올바른 입력값을 입력해주세요.")
                self.주문_처리(메인_입력)
            except Exception as e:
                print("에러 발생 : ", e)
                print("-" * 50)

    def 재고_확인(self):
        print(f'현재 재고 - 팥 : {self.붕어빵["재고"]["팥"]}개, 슈크림 : {self.붕어빵["재고"]["슈크림"]}개, 초코 : {self.붕어빵["재고"]["초코"]}개')
        print("-" * 50)

    def 주문_처리(self, 메인_입력):
        if 메인_입력[0] == "종료":
            self.메인_입력_종료()
        elif 메인_입력[0] == "관리자":
            self.메인_입력_관리자()
        else:
            self.메인_입력_붕어빵(메인_입력)

    def 메인_입력_종료(self):
        self.주문_종료 = True
        print("주문이 종료되었습니다.")
        print("-" * 50)

    def 메인_입력_관리자(self):
        관리자_비밀번호_입력 = input("관리자 비밀번호를 입력하세요. : ")
        print("-" * 50)
        self.관리자_비밀번호_확인(관리자_비밀번호_입력)

    def 관리자_비밀번호_확인(self, 관리자_비밀번호_입력):
        if 관리자_비밀번호_입력 == self.관리자_비밀번호:
            self.관리자_입력_종료 = False
            self.관리자_화면()
        else:
            self.관리자_비밀번호_불일치()

    def 관리자_화면(self):
        while not self.관리자_입력_종료:
            self.재고_확인()
            관리자_입력 = input(
                "[ --- 관리자 모드 --- ]\n"
                "추가할 붕어빵의 개수를 팥, 슈크림, 초코 순으로 「숫자만」 입력하세요.(「공백으로 구분.」 예:1 0 2 )\n"
                "관리자 모드를 종료할 경우 「종료」를 입력하세요.\n"
                "매출을 확인하려면 「매출」을 입력하세요.\n"
                "입력 : ").split(" ")
            print("-" * 50)
            try:
                if not (all(item.isdigit() for item in 관리자_입력) or len(관리자_입력) == 3 or 관리자_입력[0] in ["종료", "매출"]):
                    raise Exception("올바른 입력값을 입력해주세요.")
                self.관리자_입력_처리(관리자_입력)
            except Exception as e:
                print("에러 발생 : ", e)
                print("-" * 50)

    def 관리자_입력_처리(self, 관리자_입력):
        if 관리자_입력[0] == "종료":
            self.관리자_입력_종료 = True
            print("주문 화면으로 돌아갑니다.")
            print("-" * 50)
        elif 관리자_입력[0] == "매출":
            self.관리자_입력_매출()
        else:
            self.관리자_입력_재고_추가(관리자_입력)

    def 관리자_입력_매출(self):
        print("오늘 판매한 붕어빵의 갯수는\n"
        f'{self.붕어빵["판매_개수"]["팥"]}개, 슈크림 : {self.붕어빵["판매_개수"]["슈크림"]}개, 초코 : {self.붕어빵["판매_개수"]["초코"]}개 입니다.\n'
        f'오늘의 매출은 {self.붕어빵_매출}원입니다.')
        print("-" * 50)


    def 관리자_입력_재고_추가(self, 관리자_입력):
        self.붕어빵["추가"] = {"팥": int(관리자_입력[0]), "슈크림": int(관리자_입력[1]), "초코": int(관리자_입력[2])}
        for i in self.붕어빵["재고"].keys():
            self.붕어빵["재고"][i] += self.붕어빵["추가"][i]
        print(f'팥 붕어빵을 {self.붕어빵["추가"]["팥"]}개, 슈크림 붕어빵을 {self.붕어빵["추가"]["슈크림"]}개, 초코 붕어빵을 {self.붕어빵["추가"]["초코"]}개 추가했습니다.')
        print("-" * 50)

    def 관리자_비밀번호_불일치(self):
        print("비밀번호가 일치하지 않습니다. 처음 화면으로 돌아갑니다.")
        print("-" * 50)

    def 메인_입력_붕어빵(self, 메인_입력):
        self.붕어빵["주문"] = {"팥": int(메인_입력[0]), "슈크림": int(메인_입력[1]), "초코": int(메인_입력[2])}
        print(f'주문 내역 - 팥 : {self.붕어빵["주문"]["팥"]}개, 슈크림 : {self.붕어빵["주문"]["슈크림"]}개, 초코 : {self.붕어빵["주문"]["초코"]}개')
        print("-" * 50)
        self.주문_개수_재고_비교()

    def 주문_개수_재고_비교(self):
        for i in self.붕어빵["재고"].keys():
            if self.붕어빵["재고"][i] < self.붕어빵["주문"][i]:
                붕어빵_판매_가능_여부 = False
                break
            else:
                붕어빵_판매_가능_여부 = True
        if 붕어빵_판매_가능_여부 == True:
            self.붕어빵_재고_충분()
        else:
            self.붕어빵_재고_부족()

    def 붕어빵_재고_충분(self):
        for i in self.붕어빵["재고"].keys():
            self.붕어빵["재고"][i] -= self.붕어빵["주문"][i]
            self.붕어빵["판매_개수"][i] += self.붕어빵["주문"][i]
            self.붕어빵_매출 += self.붕어빵["가격"][i] * self.붕어빵["판매_개수"][i]
        print(f"주문이 성공적으로 완료됐습니다.")
        print("-" * 50)

    def 붕어빵_재고_부족(self):
        print(f"붕어빵의 재고가 부족합니다.")
        print("-" * 50)

    def 붕어빵_매출_저장(self):
        try:
            with open(".\bread_sell_data.csv", "x") as f:
                f.write("매출, 팥, 슈크림, 초코\n")
        except FileExistsError:
            pass
        with open(".\bread_sell_data.csv", "a") as f:
            f.write(f'{self.붕어빵_매출}, {self.붕어빵["판매_개수"]["팥"]}, {self.붕어빵["판매_개수"]["슈크림"]}, {self.붕어빵["판매_개수"]["초코"]}\n')
        print("매출이 파일에 저장되었습니다.")
        print("-" * 50)

def main():
    붕어빵_판매_시스템 = 붕어빵판매점()
    붕어빵_판매_시스템.메인_화면()
    붕어빵_판매_시스템.붕어빵_매출_저장()

if __name__ == "__main__":
    main()
```
