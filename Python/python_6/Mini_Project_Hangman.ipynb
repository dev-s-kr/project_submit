#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

7번 틀리게 되면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]

# 행맨 그림
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
]


class HangmanGame:

    def __init__(self):
        self.random_answer = random.choice(words)
        self.answer_under_bar = ["_"] * len(self.random_answer)
        self.game_over_count = len(hangman_pics) - 1
        self.now_hangman_pic = len(hangman_pics) - self.game_over_count - 1
        self.wrong_answer_count = 0
        self.user_answer = ""
        self.user_answer_right = False

    def display(self):

        while True:

            self.display_hangman_img()

            self.display_chance_count()

            self.display_answer()

            self.display_input_answer()

            self.answer_check()

            if self.game_over_count == 0:
                self.display_hangman_img()
                self.display_lose_message()
                break
            if self.wrong_answer_count == len(self.random_answer):
                self.display_hangman_img()
                self.display_win_message()
                break

    def display_hangman_img(self):
        print(hangman_pics[self.now_hangman_pic])
        self.display_line()

    def display_chance_count(self):
        print("남은 기회 :", self.game_over_count, "(0이 되면 패배입니다.)")
        self.display_line()

    def display_answer(self):
        print("단어 :", *self.answer_under_bar)
        self.display_line()

    def display_input_answer(self):
        self.user_answer = input("알파벳을 한 개만 입력하세요 : ").lower()
        self.display_line()

    def answer_check(self):
        for i in range(len(self.random_answer)):
            if self.user_answer == self.random_answer[i]:
                self.user_answer_right = True
                self.answer_under_bar[i] = self.user_answer
                self.wrong_answer_count += 1

        if self.user_answer_right:
            print(self.user_answer,"는 포함되어 있습니다.", sep = "")
            print("-" * 50)
            self.user_answer_right = False
        else:
            print(self.user_answer,"는 포함되어 있지 않습니다.", sep = "")
            print("-" * 50)
            self.game_over_count -= 1
            self.now_hangman_pic += 1

    def display_lose_message(self):
        print("정답 :", self.random_answer)
        print("게임 오버입니다.")
        self.display_line()

    def display_win_message(self):
        print("정답 :", self.random_answer)
        print("축하합니다. 승리하셨습니다.")
        self.display_line()

    def display_line(self):
        print("-" * 50)

    def play(self):
        self.display()

if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
