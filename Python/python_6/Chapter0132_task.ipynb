설명:
- address_book 리스트를 순회하며 각 연락처를 제너레이터로 반환합니다.
- 제너레이터는 메모리를 효율적으로 사용하며, 순회 시마다 하나의 연락처를 반환합니다.

```python
#가이드 코드 입니다.
# 주소록 데이터 저장을 위한 리스트
address_book = []

# 연락처 추가 함수
def add_contact(name: str, phone: str, email: str = '') -> None:
    print("--- 연락처 추가 ---")
    address_book.append(
        {
        "name": name,
        "phone": phone,
        "email": email
        }
    )
    print("연락처를 추가했습니다. :", address_book[len(address_book) - 1])
    print("-" * 50)

# 연락처 목록 보기 함수
def view_contacts() -> None:
    print("--- 연락처 목록 ---")
    if address_book:
        for address in address_book:
            print("연락처 :", address)
    else:
        print("주소록이 비어있습니다.")
    print("-" * 50)

# 연락처 검색 함수
def search_contacts(keyword: str) -> None:
    print("--- 연락처 검색 ---")
    for address in address_book:
        if address["name"] == keyword:
            print("검색 결과 :", address)
            print("-" * 50)
            return
    print("검색 결과가 없습니다.")
    print("-" * 50)

# 연락처 삭제 함수
def delete_contact(name: str) -> None:
    for address in address_book:
        if address["name"].lower() == name.lower():
            address_book.remove(address)
            view_contacts()
            print("--- 연락처 삭제 ---")
            print("삭제된 연락처 :", address)
            print("-" * 50)
            return
    print("검색 결과가 없습니다.")
    print("-" * 50)

# 연락처 정렬 함수
def sort_contacts() -> None:
    address_book.sort(key = lambda book: book["name"])
    view_contacts()
    print("연락처가 정렬되었습니다.")
    print("-" * 50)

# 제너레이터를 이용한 연락처 순회
def contact_generator():
    for address in address_book:
        yield print(address)

# 메인 함수
def main():
    add_contact("Kim", "010-1", "Kim@google")
    add_contact("Seong", "010-4", "Seong@google")
    add_contact("Park", "010-6", "Park@google")
    add_contact("Choi", "010-2", "Choi@google")
    add_contact("Lee", "010-3", "Lee@google")
    add_contact("Jeong", "010-8", "Jeong@google")
    view_contacts()
    search_contacts("Choi")
    delete_contact("choi")
    sort_contacts()
    generator = contact_generator()
    next(generator)

if __name__ == "__main__":
    main()
```

- 프로그램이 종료될 때까지 사용자 입력을 계속 받아들입니다.

```python
#가이드 코드 입니다.
# 주소록 데이터 저장을 위한 리스트
address_book = []
menu_list = ["추가", "보기", "검색", "삭제", "정렬", "종료"]

# 메뉴 선택 화면 함수
def main_menu():
    select_menu = input(
        "--- 메인 메뉴 ---\n"
        "원하는 메뉴를 괄호를 제외한 단어만 입력해주세요.\n"
        "[추가] - 연락처를 추가합니다.\n"
        "[보기] - 연락처 목록을 확인합니다.\n"
        "[검색] - 연락처를 검색합니다.\n"
        "[삭제] - 연락처를 삭제합니다.\n"
        "[정렬] - 연락처를 정렬합니다.\n"
        "[종료] - 프로그램을 종료합니다.\n"
        "입력 : "
        )
    print("-" * 50)
    try:
        check_main_input(select_menu)
        return select_menu
    except ValueError as e:
        print("에러 : ", e)
        print("-" * 50)

# 메뉴 화면 입력값 체크 함수
def check_main_input(main_input):
    if not main_input in menu_list:
        raise ValueError("올바른 입력값을 입력해주세요.")
        print("-" * 50)
    return main_input

# 연락처 정보 입력 함수
def address_input_all():
    name = input("이름을 입력하세요 : ")
    print("-" * 50)
    phone = input("번호를 입력하세요 : ")
    print("-" * 50)
    email = input("메일을 입력하세요 : ")
    print("-" * 50)
    return name, phone, email

# 이름 입력 함수
def address_input_name():
    name = input("이름을 입력하세요 : ")
    print("-" * 50)
    return name

# 연락처 추가 함수
def add_contact(name: str, phone: str, email: str = '') -> None:
    print("--- 연락처 추가 ---")
    address_book.append(
        {
        "name": name,
        "phone": phone,
        "email": email
        }
    )
    print("연락처를 추가했습니다. :", address_book[len(address_book) - 1])
    print("-" * 50)

# 연락처 목록 보기 함수
def view_contacts() -> None:
    print("--- 연락처 목록 ---")
    if address_book:
        for address in address_book:
            print("연락처 :", address)
    else:
        print("주소록이 비어있습니다.")
    print("-" * 50)

# 연락처 검색 함수
def search_contacts(keyword: str) -> None:
    print("--- 연락처 검색 ---")
    for address in address_book:
        if address["name"] == keyword:
            print("검색 결과 :", address)
            print("-" * 50)
            return
    print("검색 결과가 없습니다.")
    print("-" * 50)

# 연락처 삭제 함수
def delete_contact(name: str) -> None:
    for address in address_book:
        if address["name"].lower() == name.lower():
            address_book.remove(address)
            view_contacts()
            print("--- 연락처 삭제 ---")
            print("삭제된 연락처 :", address)
            print("-" * 50)
            return
    print("검색 결과가 없습니다.")
    print("-" * 50)

# 연락처 정렬 함수
def sort_contacts() -> None:
    address_book.sort(key = lambda book: book["name"])
    view_contacts()
    print("연락처가 정렬되었습니다.")
    print("-" * 50)

# 제너레이터를 이용한 연락처 순회
def contact_generator():
    for address in address_book:
        yield print(address)

# 프로그램을 종료하는 함수
def exit_selected():
    print("프로그램을 종료합니다.")
    print("-" * 50)

# 메인 함수
def main():
    exit_program = False
    while not exit_program:

        selected_mode = main_menu()

        if selected_mode == "추가":
            name, phone, email = address_input_all()
            add_contact(name, phone, email)

        elif selected_mode == "보기":
            view_contacts()

        elif selected_mode == "검색":
            name = address_input_name()
            search_contacts(name)

        elif selected_mode == "삭제":
            name = address_input_name()
            delete_contact(name)

        elif selected_mode == "정렬":
            sort_contacts()

        elif selected_mode == "종료":
            exit_program = True
            exit_selected()

if __name__ == "__main__":
    main()
```
