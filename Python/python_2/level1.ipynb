__[문제5]__

str = "야호야호야호호야호야야오햐오햐오햐오"

str을 3가지 이상의 다른 문자열 슬라이싱을 이용하여 확인해보세요!

출력 예)

처음 5글자: 야호야호야


```python
#변수 초기화
yaho_str =  "야호야호야호호야호야야오햐오햐오햐오"

#슬라이싱
str_slicing1 = yaho_str[0:5]
str_slicing2 = yaho_str[::-1]
str_slicing3 = yaho_str[11:2:-2]

#결과 출력
print(str_slicing1)
print(str_slicing2)
print(str_slicing3)
```

__[문제6]__

변수명 게임에 들어 있는 값 중 '홀'만 출력되도록 코드를 작성해 주세요

game = "홀짝홀짝홀짝"

출력 예시 : "홀홀홀"

<br>

__Tips__
- 문자열 슬라이싱을 이용해보세요 !


```python
#변수 초기화
game = "홀짝홀짝홀짝"

#'홀'만 출력
game_odd = game[0::2]

#결과 출력
print(game_odd)
```

__[문제7]__

변수명 전화번호에 들어 있는 번호를 모두 붙여서 출력되도록 코드를 작성해주세요.

<br>

phone = "010-1234-5678"

출력 예시 : "01012345678"

<br>

__Tips__
- split() 함수는 문자열을 특정 구분자를 기준으로 나누어 리스트로 반환합니다.


```python
#변수 초기화
phone_num = "010-1234-5678"

#번호만 표시
phone_num_split = phone_num.split("-")
just_number = phone_num_split[0] + phone_num_split[1] + phone_num_split[2]

#결과 출력
print(just_number)
```
