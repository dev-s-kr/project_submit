## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다.

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

<br>

### [동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
#마법사 능력치
wizard_int = 80

#스킬 사용 가능 여부
can_use_magic = wizard_int >= 50
print(can_use_magic)
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
#전사 능력치
warrior_str = 80
warrior_hp = 100

#스킬 사용 가능 여부
can_use_power_attack = warrior_str >= 60 and warrior_hp >= 80
print(can_use_power_attack)
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
#도적 능력치
rogue_agi = 80
rogue_str = 40

#스킬 사용 가능 여부
can_use_dodge = rogue_agi >= 70 or rogue_str >= 50
print(can_use_dodge)
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
#힐러 능력치
healer_int = 70
healer_hp = 70

#스킬 사용 가능 여부
can_use_heal = healer_int >= 60 and healer_hp >= 50
print(can_use_heal)
```

<br>

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
#레벨업에 필요한 경험치
level_max_exp = 100

#각 캐릭터의 경험치
wizard_exp = 120
warrior_exp = 150
rogue_exp = 110
healer_exp = 90

#레벨업 여부 판단
can_level_up = wizard_exp >= level_max_exp
print(can_level_up)

can_level_up = warrior_exp >= level_max_exp
print(can_level_up)

can_level_up = rogue_exp >= level_max_exp
print(can_level_up)

can_level_up = healer_exp >= level_max_exp
print(can_level_up)
```
