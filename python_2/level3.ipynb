__[문제1]__

미션 부수기 챌린지

다음 미션을 차례대로 부숴주세요.

<br>

__미션__
- x에 5를 할당하고, y에는 10을 할당
- x의 값을 제곱하고, 그 값을 z에 할당
- y의 값을 2로 나눈 후, 그 값을 a에 할당
- a와 z를 더한 값을 b에 할당
- b의 값이 x와 같은지 출력
-  b와 a를 문자열로 변환하여 합친 결과를 출력


<br>

```python
#변수 초기화
x = 5
y = 10
z = x ** 2
a = y / 2
b = a + z

#값 비교
b_equals_x = b == x

#문자열로 변환하여 합치기
sum_str_b_a = str(b) + str(a)

#결과 출력
print(b_equals_x)
print(sum_str_b_a)
```

이 돌다리는 반드시 두드려 보고 건너야 합니다.

미션을 해결하면 자동으로 단단한 돌다리인지 확인 됩니다!

__[문제6]__

창의성, 시장성, 안정성, 수익성에 대한 각각의 평가 중 아래 기준에 부합하지 못하면 창업지원을 받을 수 없습니다.

- 수익성 60점 이상
- 창의성과 시장성 50점 이상
- 시장성과 수익성을 더한 값의 절반이 안정성 보다 높아야 함


다음 점수가 입력되었을때 창업 지원이 가능한지 여부를 True 또는 False로 출력해주세요.

---

창의성 : 62
시장성 : 71
안정성 : 68
수익성 : 66
<br>

* 참고
(creativity ,
marketability ,
stability ,
profitability )

<br>

__Tips__
* 비교 연산자와 논리 연산자의 결과를 고려해보아요


```python
#변수 초기화
creativity_score = 62
marketability_score = 71
stability_score = 68
profitability_score = 66

#조건
standard1 = profitability_score >= 60
standard2 = creativity_score >= 50 and marketability_score >= 50
standard3 = (marketability_score + profitability_score) / 2 > stability_score

#조건 전부 만족하는지 검사
pass_result = standard1 and standard2 and standard3

#결과 출력
print(pass_result)
```
