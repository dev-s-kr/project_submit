__[문제6]__


오즈 코딩 스쿨 백엔드 8기 과정에 참여한 수강생 중 5명의 수강생이 자율적으로 스터디 그룹을 결성했습니다.

스터디 그룹 내에는 한 주간 배운 내용으로 매주 쪽지 시험을 보는 룰이 있습니다.

이 쪽지 시험의 평균이 70점이 넘으면 주말 스터디는 없습니다.

주말 스터디의 여부를 True 또는 False로 출력해주세요.

<br>
A: 70, B: 75, C: 50, D: 80, E: 60

```python
#변수 초기화
student_num = 5
pass_average = 70
student_a = 70
student_b = 75
student_c = 50
student_d = 80
student_e = 60

#평균 계산 및 비교
score_sum = student_a + student_b + student_c + student_d + student_e
score_average = score_sum / 5
pass_weekend_study = score_average >= pass_average

#결과 출력
print(pass_weekend_study)
```
